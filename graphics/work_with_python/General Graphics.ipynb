### Bar plot

```python
def plot_bar_rows_columns(x,y):
    normdata = colors.Normalize(min(y), max(y))
    colormap = cm.get_cmap("Oranges")
    colores =colormap(normdata(y))
    edgecolor = "#2A0A0A"
    fig, ax = plt.subplots(figsize=(15, 9))
    rects = ax.bar(x, y, color=colores, edgecolor=edgecolor)
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  
                    textcoords="offset points",
                    ha='center', va='bottom')
    ax.set_ylabel('Numero de Filas')
    ax.set_xlabel('Columnas')
    ax.set_title('Cantidad de valores validos por columnas')
    labels = ax.get_xticklabels()
    plt.setp(labels, rotation=45, horizontalalignment='right')
    plt.show()
```
